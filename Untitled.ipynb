```python
def get_var(file, evil = set()):
    cont = set()
    pairs = set()
    for line in f:
        a, b, c = line.split()
        if a in evil or b in evil:
            continue
        cont.add(a)
        cont.add(b)
        pair = frozenset({a,b})
        pairs.add(pair)
    return cont, pairs

def get_update_dist(file):
    cont_len = {}
    for line in file:
        cont, length = line.split()
        length = int(length)
        if cont in cont_len:
            print("Error: double cont: " + cont + " | v1: " + str(length) + " | v2: " + str(cont_len[cont]))
            if length < 5:
                print("ignore: " + str(length))
                continue
            print("ignore: " + str(cont_len[cont]))
            #raise
        cont_len[cont] = length
    return cont_len

def Contig_std(var, pairs, cont_len, file, evil = set()):
    print("Löse LP")

    # Setup Model
    model = Model("Gen")
    #model.setParam("TimeLimit",30)
    #model.setParam(GRB.Param.LogToConsole, 0.0)
    model.Params.OptimalityTol = 1e-3
    # Variablen
    print("setze Variablen")
    x = {}
    for n in var:
        if n in evil:
            continue
        x[n] = model.addVar(vtype=GRB.CONTINUOUS, lb=0.0, name = n)
    eps = {}
    for pair in pairs:
        c = list(pair)
        if c[0] in evil or c[1] in evil:
            continue
        eps[pair] = model.addVar(vtype=GRB.CONTINUOUS, name = c[0] + " <-> " + c[1])
    # Kreis-Nebenbedingungen
    print("Generating constraints ...")
    for line in file:
        a, b, dist = line.split()
        if a in evil or b in evil:
            continue
        dist = float(dist) + cont_len[a]
        e = frozenset({a,b})
        model.addConstr(x[b] - x[a]  - dist <= eps[e])
        model.addConstr(-x[b] + x[a]  + dist <= eps[e])

    # Lösen
    print("Solving instances ...")
    model.setObjective(sum(eps.values())/len(eps), GRB.MINIMIZE)
    model.update()
    model.optimize()
    return model


def Contig_rem_bad(var, pairs, cont_len, file):
    print("Löse LP")

    # Setup Model
    model = Model("Gen")
    #model.setParam("TimeLimit",30)
    #model.setParam(GRB.Param.LogToConsole, 0.0)
    #model.Params.OptimalityTol = 1e-3
    # Variablen
    print("setze Variablen")
    x = {}
    for n in var:
        x[n] = model.addVar(vtype=GRB.CONTINUOUS, lb=0.0, name = n)
    eps = {}
    to_bad = {}
    for pair in pairs:
        c = list(pair)
        eps[pair] = model.addVar(vtype=GRB.CONTINUOUS, name = c[0] + " ---> " + c[1])
        to_bad[pair] = model.addVar(vtype=GRB.BINARY, name = c[0] + " -/-> " + c[1])
    # Kreis-Nebenbedingungen
    print("Generating constraints ...")
    for line in file:
        a, b, dist = line.split()
        dist = float(dist) + cont_len[a] -1
        e = frozenset({a,b})
        model.addConstr( x[b] - x[a]  - dist <= eps[e] + 1000000 * to_bad[e])
        model.addConstr(-x[b] + x[a]  + dist <= eps[e] + 1000000 * to_bad[e])

    # Lösen
    print("Solving instances ...")
    model.setObjective((sum(eps.values()) + 100 * sum(to_bad.values()))/len(eps), GRB.MINIMIZE)
    model.update()
    model.optimize()
    return model

def Contig_rem_bad_var(var, pairs, cont_len, file, evil = set()):
    print("Löse LP")

    # Setup Model
    model = Model("Gen")
    #model.setParam("TimeLimit",30)
    #model.setParam(GRB.Param.LogToConsole, 0.0)
    #model.Params.OptimalityTol = 1e-3
    # Variablen
    print("setze Variablen")
    x = {}
    to_bad = {}
    for n in var:
        x[n] = model.addVar(vtype=GRB.CONTINUOUS, lb=0.0, name = n)
        to_bad[n] = model.addVar(vtype=GRB.BINARY, name = 'not ' + n)
    eps = {}
    for pair in pairs:
        c = list(pair)
        eps[pair] = model.addVar(vtype=GRB.CONTINUOUS, name = c[0] + " ---> " + c[1])
    # Kreis-Nebenbedingungen
    print("Generating constraints ...")
    for line in file:
        a, b, dist = line.split()
        if a in evil or b in evil:
            continue
        dist = float(dist) + cont_len[a] -1
        e = frozenset({a,b})
        model.addConstr( x[b] - x[a]  - dist <= eps[e] + 1000000 * to_bad[a] + 1000000 * to_bad[b])
        model.addConstr(-x[b] + x[a]  + dist <= eps[e] + 1000000 * to_bad[a] + 1000000 * to_bad[b])

    # Lösen
    print("Solving instances ...")
    model.setObjective((sum(eps.values()) + 10000 * sum(to_bad.values()))/len(eps), GRB.MINIMIZE)
    model.update()
    model.optimize()
    return model
```
